```python
import pandas as pd
import statistics
import numpy as np
from statistics import mean


#exp1_df
exp2_df = pd.read_csv("Experiment2.csv")
exp3_df = pd.read_csv("Experiment3.csv")
exp4_df = pd.read_csv("Experiment4.csv")
```

```python
rf_means = []
svm_means = []
knn_means = []
```

```python
exp2_df
```

```python
rf_acc_arr_delta = []
svm_acc_arr_delta = []
knn_acc_arr_delta = []
rf_acc_arr_theta = []
svm_acc_arr_theta = []
knn_acc_arr_theta = []
rf_acc_arr_alpha = []
svm_acc_arr_alpha = []
knn_acc_arr_alpha = []
rf_acc_arr_beta = []
svm_acc_arr_beta = []
knn_acc_arr_beta = []
rf_acc_arr_gamma = []
svm_acc_arr_gamma = []
knn_acc_arr_gamma = []

exp2_acc = exp2_df['Accuracy']

print(len(exp2_acc))
for i in range (0,len(exp2_acc)):
    if (i%15 == 0):
        rf_acc_arr_delta.append(exp2_acc[i])
    if (i%15 == 1):
        svm_acc_arr_delta.append(exp2_acc[i])
    if (i%15 == 2):
        knn_acc_arr_delta.append(exp2_acc[i])
    if (i%15 == 3):
        rf_acc_arr_theta.append(exp2_acc[i])
    if (i%15 == 4):
        svm_acc_arr_theta.append(exp2_acc[i])
    if (i%15 == 5):
        knn_acc_arr_theta.append(exp2_acc[i])
    if (i%15 == 6):
        rf_acc_arr_alpha.append(exp2_acc[i])
    if (i%15 == 7):
        svm_acc_arr_alpha.append(exp2_acc[i])
    if (i%15 == 8):
        knn_acc_arr_alpha.append(exp2_acc[i])
    if (i%15 == 9):
        rf_acc_arr_beta.append(exp2_acc[i])
    if (i%15 == 10):
        svm_acc_arr_beta.append(exp2_acc[i])
    if (i%15 == 11):
        knn_acc_arr_beta.append(exp2_acc[i])
    if (i%15 == 12):
        rf_acc_arr_gamma.append(exp2_acc[i])
    if (i%15 == 13):
        svm_acc_arr_gamma.append(exp2_acc[i])
    if (i%15 == 14):
        knn_acc_arr_gamma.append(exp2_acc[i])
```

```python
rf_mean_delta = statistics.mean(rf_acc_arr_delta)
print("RF mean delta: ", rf_mean_delta)
svm_mean_delta = statistics.mean(svm_acc_arr_delta)
print("SVM mean delta: ", svm_mean_delta)
knn_mean_delta = statistics.mean(knn_acc_arr_delta)
print("kNN mean delta: ",knn_mean_delta)

rf_mean_theta = statistics.mean(rf_acc_arr_theta)
print("RF mean theta: ", rf_mean_theta)
svm_mean_theta = statistics.mean(svm_acc_arr_theta)
print("SVM mean theta: ", svm_mean_theta)
knn_mean_theta = statistics.mean(knn_acc_arr_theta)
print("kNN mean theta: ",knn_mean_theta)

rf_mean_alpha = statistics.mean(rf_acc_arr_alpha)
print("RF mean alpha: ", rf_mean_alpha)
svm_mean_alpha = statistics.mean(svm_acc_arr_alpha)
print("SVM mean alpha: ", svm_mean_alpha)
knn_mean_alpha = statistics.mean(knn_acc_arr_alpha)
print("kNN mean alpha: ",knn_mean_alpha)

rf_mean_beta = statistics.mean(rf_acc_arr_beta)
print("RF mean beta: ", rf_mean_beta)
svm_mean_beta = statistics.mean(svm_acc_arr_beta)
print("SVM mean beta: ", svm_mean_beta)
knn_mean_beta = statistics.mean(knn_acc_arr_beta)
print("kNN mean beta: ",knn_mean_beta)

rf_mean_gamma = statistics.mean(rf_acc_arr_gamma)
print("RF mean gamma: ", rf_mean_gamma)
svm_mean_gamma = statistics.mean(svm_acc_arr_gamma)
print("SVM mean gamma: ", svm_mean_gamma)
knn_mean_gamma = statistics.mean(knn_acc_arr_gamma)
print("kNN mean gamma: ",knn_mean_gamma)

rf_means.append(rf_mean_delta)
rf_means.append(rf_mean_theta)
rf_means.append(rf_mean_alpha)
rf_means.append(rf_mean_beta)
rf_means.append(rf_mean_gamma)

svm_means.append(svm_mean_delta)
svm_means.append(svm_mean_theta)
svm_means.append(svm_mean_alpha)
svm_means.append(svm_mean_beta)
svm_means.append(svm_mean_gamma)

knn_means.append(knn_mean_delta)
knn_means.append(knn_mean_theta)
knn_means.append(knn_mean_alpha)
knn_means.append(knn_mean_beta)
knn_means.append(knn_mean_gamma)
```

```python
rf_lit_mean = statistics.mean(rf_lit_acc_arr)
print("RF literature mean: ", rf_lit_mean)
svm_lit_mean = statistics.mean(svm_lit_acc_arr)
print("SVM literature mean: ", svm_lit_mean)
rf_own_mean = statistics.mean(rf_own_acc_arr)
print("RF own mean: ", rf_own_mean)
svm_own_mean = statistics.mean(svm_own_acc_arr)
print("SVM own mean: ", svm_own_mean)

rf_means.append(rf_own_mean)
rf_means.append(rf_lit_mean)

svm_means.append(svm_own_mean)
svm_means.append(svm_lit_mean)

knn_means.append(None)
knn_means.append(None)
```
