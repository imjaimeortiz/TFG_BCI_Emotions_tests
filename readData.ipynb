```python
exp3_acc = exp3_df.sort_values(by="Accuracy", ascending=True)
exp3_acc = exp3_acc['Accuracy']
values = exp3_acc.values
values
```

```python
rf_acc_arr = []
svm_acc_arr = []
knn_acc_arr = []

print(len(exp3_acc))
for i in range (0,len(exp3_acc)):
    if (i%3 == 0):
        rf_acc_arr.append(exp3_acc[i])
    if (i%3 == 1):
        svm_acc_arr.append(exp3_acc[i])
    if (i%3 == 2):
        knn_acc_arr.append(exp3_acc[i])

print(rf_acc_arr)
print(svm_acc_arr)
print(knn_acc_arr)
```

```python
print("RF mean: ", statistics.mean(rf_acc_arr))
print("SVM mean: ", statistics.mean(svm_acc_arr))
print("kNN mean: ", statistics.mean(knn_acc_arr))
```

```python
exp4_acc = exp4_df.sort_values(by="Accuracy", ascending=True)
exp4_acc = exp4_acc['Accuracy']
values = exp4_acc.values
values
```

```python
rf_lit_acc_arr = []
svm_lit_acc_arr = []
rf_own_acc_arr = []
svm_own_acc_arr = []

print(len(exp4_acc))
for i in range (0,len(exp4_acc)):
    if (i%4 == 0):
        rf_lit_acc_arr.append(exp4_acc[i])
    if (i%4 == 1):
        svm_lit_acc_arr.append(exp4_acc[i])
    if (i%4 == 2):
        rf_own_acc_arr.append(exp4_acc[i])
    if (i%4 == 3):
        svm_own_acc_arr.append(exp4_acc[i])


print(rf_lit_acc_arr)
print(svm_lit_acc_arr)
print(rf_own_acc_arr)
print(svm_own_acc_arr)
```

```python
print("RF literature mean: ", statistics.mean(rf_lit_acc_arr))
print("SVM literature mean: ", statistics.mean(svm_lit_acc_arr))
print("RF own mean: ", statistics.mean(rf_own_acc_arr))
print("SVM own mean: ", statistics.mean(svm_own_acc_arr))
```
